### Step 1 - *"Carving"* your model

Now you will:

  - (**-i**) export a model initialized with M9 as default medium

```python
!carve files1/ecoli_genome.faa -o files1/ecoli_model.xml -g M9 -i M9 -v
```

### Step 2 - Loading your model

In this part of the tutorial we will use a simulation library called **fluxbound**. 

> ⚠️ First take a moment to quickly check [the documentation](https://cdanielmachado.github.io/fluxbound/).

Now let's load our model and see some basic info:

```python
from fluxbound import load_model
model = load_model('files1/ecoli.xml')
model
```

### Step 3 - Flux Balance Analysis

Let's run a very simple FBA simulation:

```python
from fluxbound import FBA
solution = FBA(model)
print(solution)
```

```python
solution.show_values(pattern='R_EX', sort=True, abstol=0.1)
```

We can easily change the growth conditions. 

Let's simulate anaerobic growth:

```python
FBA(model, constraints={'R_EX_o2_e': 0})
```
