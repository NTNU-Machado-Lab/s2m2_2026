# Welcome 🙂

This tutorial consists of two parts:


### Part 1 - GEM reconstruction and analysis

👉 [**Here**](1_CarveMe.ipynb) you will learn how to use **CarveMe** to build a genome-scale metabolic model. 
You will then load the model and analyse it using a some simple methods.


### Part 2 - Microbial community simulation

👉 [**Here**](2_SteadierCom.ipynb) you will learn how to use **SteadierCom** to simulate a microbial community. 
You will then analyse the results and look at cross-feeding interactions between microbes.
